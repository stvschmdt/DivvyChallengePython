```python
#Divvy Bike Challenge (Homework)
#USER TOGGLE below allows to run the data below for various times/days of week
#Days are Mon-Fri 1-5, Sat 6, Sun 7
#DEFAULT Setting is least data - Sunday Overnight
#This should be run in ipython notebook with flag --pylab (inline optional)
#The MAP CALL can also be modified by user to show various regions in 'zoom' or modify titles


from mpl_toolkits.basemap import Basemap
from netCDF4 import Dataset, date2index
import numpy as np
import pandas as pd
from pandas import DataFrame, Series
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
trips = pd.read_csv('Divvy_Trips_2013.csv',
                 parse_dates=[1, 2], low_memory=False)
stations = pd.read_csv('Divvy_Stations_2013.csv')
```

```python
trips['starttime'] = pd.to_datetime(trips['starttime'])
```

```python
trips['starthour'] = trips.apply((lambda row: row['starttime'].hour), axis=1)
```

```python
trips['startdayofweek']= trips.apply((lambda row: row['starttime'].isoweekday()), axis=1)
```

```python
###USER TOGGLE###

#trips = trips[trips['startdayofweek'] < 6]
#trips = trips[trips['startdayofweek'] == 6]
trips = trips[trips['startdayofweek'] == 7]
trips = trips[(trips['starthour'] > 5) | (trips.starthour <=13)]
#trips = trips[(trips.starthour > 13) | (trips.starthour <=21) ]
#trips = trips[(trips.starthour <=5) | (trips.starthour >21)]
```

```python
trips.count()
```

```python
trip_summary = trips.groupby(["from_station_name", "to_station_name"])['tripduration'].agg({'number_trips' : size, 'avg_duration' : np.mean})
trip_summary = trip_summary.reset_index()
```

```python
stations_from =stations[['name', 'latitude', 'longitude']]
stations_from.columns = ['from_station_name', 'from_latitude', 'from_longitude']
stations_to =stations[['name', 'latitude', 'longitude']]
stations_to.columns = ['to_station_name', 'to_latitude', 'to_longitude']
trip_summary = pd.merge(trip_summary, stations_from)
trip_summary = pd.merge(trip_summary, stations_to)
```

```python
trip_summary.head()
```

```python
trip_summary["distance"] = 0
for i in range(0, len(trip_summary)):
    trip_summary["distance"].ix[i] = sqrt((trip_summary["from_latitude"].ix[i]-trip_summary["to_latitude"].ix[i])**2 + (trip_summary["from_longitude"].ix[i]-trip_summary["to_longitude"].ix[i])**2)
```

```python
trip_summary.head()
```

```python
trip_summary.drop(["to_station_name", "avg_duration"], axis=1)
```

```python
stats = trip_summary.groupby(["from_station_name", "from_latitude", "from_longitude"]).sum()
```

```python

stats["ave_dist"] = 0
for i in range(0,len(stats)):
    stats.ave_dist.ix[i] = stats.distance.ix[i] / stats.number_trips.ix[i]
stats["colorgroup"] = 0
for i in range(0,len(stats)):
               stats.colorgroup.ix[i] = round(100*stats.ave_dist.ix[i]/(stats.ave_dist.max() - stats.ave_dist.min()))
stats.columns
```

```python
stats.head()
```

```python
stats.index[0], stats.colorgroup.ix[0]
toplotx = []
toploty = []
toplotz = []
for i in range(0,len(stats)):
    toplotx.append(stats.index[i][1])
    toploty.append(stats.index[i][2])
    toplotz.append(stats.colorgroup.ix[i])
```

```python
data = pd.DataFrame(toplotx, columns=['x'])
data['y'] = toploty
data['z'] = toplotz
```

```python
def get_marker_color(magnitude):
    # b: blue
#g: green
#r: red
#c: cyan
#m: magenta
#y: yellow
#k: black
#w: white
    if magnitude < 10.0:
        return ('wo')
    elif magnitude >= 10 and magnitude < 30:
        return ('yo')
    elif magnitude >= 30 and magnitude < 50:
        return ('go')
    elif magnitude >= 50 and magnitude < 70:
        return ('mo')
    else:
        return ('ro')
```

```python
# MAP CALL
# Corners of the map
llcrnrlat = stations['latitude'].min()
llcrnrlon = stations['longitude'].min()
urcrnrlat = stations['latitude'].max()
urcrnrlon = stations['longitude'].max()

# Adjust corners
urcrnrlat = 41.950000  # North Chicago
llcrnrlat = 41.786033  # 55th ish

plt.figure(figsize=(7, 8))
plt.title("Distance Traveled From Originating Station\nLEAST | White -> Yellow -> Magenta -> Red | GREATEST\n")



#The basemap object
m = Basemap(llcrnrlon=llcrnrlon,
            llcrnrlat=llcrnrlat,
            urcrnrlon=urcrnrlon,
            urcrnrlat=urcrnrlat,
            lon_0=-87.632446, lat_0=41.872884,
            # Lake Shore Dr and Chicago River is the center
            area_thresh=0.001,  # seems to have no effect
            projection='merc', resolution='i') 

m.drawcoastlines()
# The resolution of the above for Chicago is very poor but shows the effect.

m.drawrivers()  # Doesn't draw the Chicago river
m.drawmapboundary(fill_color='aqua')
m.fillcontinents(color='coral', lake_color='skyblue', alpha=.8)

# The stations
#m.plot(stations.longitude.values, stations.latitude.values, 'bo', latlon=True)
#m.plot(data.y.values, data.x.values, marker='o', get_marker_color(data.z.values), alpha=.7, latlon=True)
for i in range(0,300):
    this = get_marker_color(data.z.ix[i])
    m.plot(data.y.ix[i], data.x.ix[i], this, marker='o', alpha=.7, latlon=True)
# The shapefiles

m.readshapefile('./shapefiles/landmarks/landmarks_geo',
                'landmarks', linewidth=1, color='k')
m.readshapefile('./shapefiles/parks/parks_geo',
                'parks', linewidth=.5, color='k')
m.readshapefile('./shapefiles/majorstreets/major_streets_geo',
                'majorstreets', linewidth=.5, color='k')
#m.readshapefile('./shapefiles/bike_routes/bike_routes_geo',
 #               'bikeroutes', linewidth=1.0, color='b')
pylab.ion()
```
